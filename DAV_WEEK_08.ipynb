```python
# Import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# -----------------------------
# Step 1: Create sample dataset
# -----------------------------
np.random.seed(42)  # for reproducibility

# Simulate data for 12 months
data = {
    "Monthly_Sales": np.random.normal(20000, 3000, 12),
    "Ad_Spend": np.random.normal(5000, 1000, 12),
    "Discount": np.random.uniform(5, 20, 12),
    "Customer_Visits": np.random.normal(1000, 200, 12),
    "Online_Reviews": np.random.normal(200, 50, 12)
}

df = pd.DataFrame(data)

# Optional: Create some realistic relationships
df["Monthly_Sales"] += 2.5 * df["Ad_Spend"] - 150 * df["Discount"] + np.random.normal(0, 1000, 12)

# -----------------------------
# Step 2: Compute correlations
# -----------------------------
corr_matrix = df.corr()

print("=== Correlation Matrix ===")
print(corr_matrix)

# -----------------------------
# Step 3: Plot heatmap
# -----------------------------
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Retail Variables")
plt.show()

# -----------------------------
# Step 4: Compare correlations
# -----------------------------
sales_ad_corr = corr_matrix.loc["Monthly_Sales", "Ad_Spend"]
sales_discount_corr = corr_matrix.loc["Monthly_Sales", "Discount"]

print(f"\nCorrelation between Sales and Ad Spend: {sales_ad_corr:.2f}")
print(f"Correlation between Sales and Discount: {sales_discount_corr:.2f}")

if abs(sales_ad_corr) > abs(sales_discount_corr):
    print("👉 Advertising Spend has a stronger correlation with Sales.")
elif abs(sales_ad_corr) < abs(sales_discount_corr):
    print("👉 Discounts have a stronger correlation with Sales.")
else:
    print("👉 Advertising Spend and Discounts have an equal correlation with Sales.")
```
