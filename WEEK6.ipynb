```python
# Install required libraries if not already installed
# !pip install seaborn matplotlib numpy pandas

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ---- Step 1: Simulate blood sugar levels for 200 patients ----
# 120 normal patients (mean=100, std=10)
# 80 diabetic patients (mean=160, std=15)

np.random.seed(42)
normal_patients = np.random.normal(100, 10, 120)
diabetic_patients = np.random.normal(160, 15, 80)
blood_sugar = np.concatenate([normal_patients, diabetic_patients])

# ---- Step 2: Create a dataframe ----
df = pd.DataFrame({'Blood_Sugar': blood_sugar})

# ---- Step 3: Plot KDE (Density Plot) ----
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x='Blood_Sugar', fill=True, color='crimson', bw_adjust=0.5)

# ---- Step 4: Add labels, title, and grid ----
plt.title('Distribution of Blood Sugar Levels Among 200 Patients', fontsize=14)
plt.xlabel('Blood Sugar Level (mg/dL)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(True)
plt.show()
```
